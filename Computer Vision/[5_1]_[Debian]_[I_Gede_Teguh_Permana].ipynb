```python
import cv2 as cv
from skimage import io
from google.colab.patches import cv2_imshow # for image display
import matplotlib.pyplot as plt
```

```python
img_path = 'https://media.istockphoto.com/photos/coins-collection-isolated-on-white-picture-id147063335?k=20&m=147063335&s=612x612&w=0&h=KLZ8GF9AjbjoK4OXTcZjnZDogT-hV6Hr7atjczfFhTI='
```

### Read and visualize the image and then save it to a variable

```python
# Visualize the image background
image_bgr = io.imread(img_path)
cv2_imshow(image_bgr)
```

```python
# Visualize the rgb image
image_rgb = cv.cvtColor(image_bgr,cv.COLOR_BGR2RGB)
cv2_imshow(image_rgb)
```

# Convert to gray scale

```python
image_gray = cv.cvtColor(image_rgb, cv.COLOR_RGB2GRAY)
cv2_imshow(image_gray)
```

### Show the histogram of the grayscale image

```python
plt.hist(image_gray.ravel(),bins = 256, range = [0,256]) 
plt.show()
```

# threshold value

```python
ret, thresh_hold = cv.threshold(image_gray, 100, 255, cv.THRESH_BINARY) 
ret, thresh_hold1 = cv.threshold(image_gray, 100, 255, cv.THRESH_BINARY_INV) 
ret, thresh_hold2 = cv.threshold(image_gray, 100, 255, cv.THRESH_TOZERO) 
ret, thresh_hold3 = cv.threshold(image_gray, 100, 255, cv.THRESH_TOZERO_INV) 
ret, thresh_hold4 = cv.threshold(image_gray, 100, 255, cv.THRESH_TRUNC)   
 
thresh_hold = cv.resize(thresh_hold, (960, 540))    
cv2_imshow(thresh_hold) 
 
thresh_hold1 = cv.resize(thresh_hold1, (960, 540))    
cv2_imshow(thresh_hold1) 
 
thresh_hold2 = cv.resize(thresh_hold2, (960, 540))    
cv2_imshow(thresh_hold2) 
 
thresh_hold3 = cv.resize(thresh_hold3, (960, 540))    
cv2_imshow(thresh_hold3) 
 
thresh_hold4= cv.resize(thresh_hold4, (960, 540))    
cv2_imshow(thresh_hold4) 
 
if cv.waitKey(0) & 0xff == 25:  
    cv.destroyAllWindows()
```

```python
mask = cv.bitwise_not(thresh_hold)
mask1 = cv.bitwise_not(thresh_hold1)
mask2 = cv.bitwise_not(thresh_hold2)
mask3 = cv.bitwise_not(thresh_hold3)
mask4 = cv.bitwise_not(thresh_hold4)

cv2_imshow(mask)
cv2_imshow(mask1) 
cv2_imshow(mask2) 
cv2_imshow(mask3)
cv2_imshow(mask4)   

if cv.waitKey(0) & 0xff == 25:  
    cv.destroyAllWindows()
```
